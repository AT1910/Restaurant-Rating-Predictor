```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split                
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)       
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
# Getting the function ready
def modelresults(predictions):
    print("Mean absolute error on model is {}".format(mean_absolute_error(Y_test, predictions)))
    print("Root mean squared error on model is {}".format(np.sqrt(mean_squared_error(Y_test, predictions))))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
lr.predict(X_test)
```

```python
predictionsfromlr = lr.predict(X_test)
```

```python
modelresults(predictionsfromlr)
```

```python
predictionsfromlr
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
svrmodel = SVR()
```

```python
#Dictionary for parameters, you can decrease the number of parameters to speed up the process
param_gridsvr = {"C": [0.1, 0.5], "kernel": ['linear', "poly"], "degree": [2]}
```

```python
# Training our model using GridSearchCV
gridsvr = GridSearchCV(svrmodel, param_gridsvr)
gridsvr.fit(X_train, Y_train)
```

```python
gridsvr.predict(X_test)
```

```python
predsgridsvr = gridsvr.predict(X_test)
```

```python
modelresults(predsgridsvr)
```

```python
# Train a tree model
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid = {
    "max_depth": [10, 30],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2]
}
```

```python
treemodel = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator = treemodel, param_grid = param_grid, cv=5) #Cross validation
```

```python
grid_tree.fit(X_train, Y_train)
```

```python
grid_tree.predict(X_test)
```

```python
treepredictions = grid_tree.predict(X_test)
```

```python
modelresults(treepredictions)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfmodel = RandomForestRegressor()
```

```python
param_gridfr = {"max_depth": [5 , 15],"n_estimators": [2, 5, 10]}
```

```python
gridrfr = GridSearchCV(estimator=rfmodel, param_grid=param_gridfr)
```

```python
gridrfr.fit(X_train, Y_train)
```

```python
randomforestpredictions = gridrfr.predict(X_test)
```

```python
modelresults(randomforestpredictions)
```

```python
# What if you want to see the best performing parameters 
gridrfr.best_params_
```

```python
grid_tree.best_params_
```
